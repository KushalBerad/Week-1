```python
img = cv2.imread("dataset/sample_car.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,5))
plt.imshow(img_rgb)
plt.title("Sample Automotive Image - Car Detection")
plt.axis("off")
plt.show()
```

```python
classes = ["car", "truck", "bus", "motorbike", "pedestrian"]
num_images = 1500
print(f"Dataset Summary:")
print(f"Total Images (approx.): {num_images}")
print(f"Classes Detected: {classes}")
```

```python
x, y, w, h = 130, 150, 250, 100
img_box = img_rgb.copy()
cv2.rectangle(img_box, (x, y), (x+w, y+h), (255, 0, 0), 3)
cv2.putText(img_box, "car", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

plt.figure(figsize=(8,5))
plt.imshow(img_box)
plt.title("Example - Object Detection Bounding Box")
plt.axis("off")
plt.show()
```

```python
objects = {"car": 500, "truck": 300, "bus": 200, "motorbike": 250, "pedestrian": 250}
plt.bar(objects.keys(), objects.values(), color='skyblue')
plt.title("Class Distribution (Example)")
plt.xlabel("Object Type")
plt.ylabel("Count")
plt.show()
```
